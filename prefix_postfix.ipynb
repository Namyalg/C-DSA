```
#POSTFIX

#in this case operand popped out first will be operand2, then comes operand1

#for prefix it is reverse


exp = "4 55 + 62 23 - *"
exp = exp.split()
stack = []
#print(exp)
for i in range(len(exp)):
  #if exp[i] == "+" or exp[i] == "-" or exp[i] == "*" or exp[i] == "/":
  if exp[i] in "+-*/^":
    op2 = stack.pop()
    op1 = stack.pop()
    if exp[i] == "+":
      res = op1 + op2
    elif exp[i] == "-":
      res = op1 - op2
    elif exp[i] == "*":
      res = op1 * op2
    elif exp[i] == "/":
      res = op1 / op2
    elif exp[i] == "^":
      res = op1 ^ op2
    stack.append(res)
  else:
    stack.append(int(exp[i]))

print(stack[0])
```

```
def binary_search(arr, x): 
    low = 0
    high = len(arr) - 1
    mid = 0
  
    while low <= high: 
  
        mid = (high + low) // 2
  
 
        if arr[mid] < x: 
            low = mid + 1
  
        # If x is greater, ignore left half 
        elif arr[mid] > x: 
            high = mid - 1
  
        # If x is smaller, ignore right half 
        else: 
            return mid 
  
    # If we reach here, then the element was not present 
    return -1
  
  
# Test array 
arr = [-283, 2, 27, 39, 83, 282, 372, 839]
x = 100
  
# Function call 
result = binary_search(arr, x) 
print(result)
```

```
def d(a,n):
  if 0 not in a:
    return(0)
  else:
    e = min(a)
    f = max(a)
    if [*range(e, f+1)] == a:
      return(0)
    else:
      a = sorted(a)
      p1 = a[:n//2]
      p2 = a[n//2:]
      s = 0
      for i in range(p1[0], p1[-1]):
        if i not in p1:
          s += i
          break
      for i in range(p2[0] , p2[-1]):
        if i not in p2:
          s += i
      return(s)
  
for _ in range(int(input())):
  n = int(input())
  a = list(map(int, input().split()))
  print(d(a,n))
```
